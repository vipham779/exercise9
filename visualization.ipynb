```python
# 2
# V·∫Ω ƒë·ªì th·ªã h√†m sin,cos,tan trong kho·∫£ng -2ùúã -> 2ùúã
```

```python
import matplotlib.pyplot as plt
x = np.arange(-2*np.pi, 2*np.pi+0.1, 0.1)
y = np.sin(x)
z = np.cos(x)
t = np.tan(x)
plt.plot(x, y)
plt.plot(x, z)
plt.plot(x, t)
plt.xlabel('x values from -2pi to 2pi')
plt.ylabel('sin(x), cos(x) and tan(x)')
plt.legend(['y = sin(x)', 'y = cos(x)', 'y = tan(x)'])
```
